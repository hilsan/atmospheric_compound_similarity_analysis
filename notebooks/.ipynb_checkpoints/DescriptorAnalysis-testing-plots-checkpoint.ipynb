```python
import joblib as jl 
import pandas as pd
import rdkit; 
import sys, os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
ROOT_DIR = os.path.dirname(os.path.abspath('/scratch/work/sandsth2/SCRIPTS/MolAnalysis'))
sys.path.append(ROOT_DIR)
from MolAnalysis.descriptors import *
from MolAnalysis.loaddata import *
from MolAnalysis.compileanalysis import *
```

```python
def load_datasets():
    """Load dataset locations. Update to filelist?"""
    datasets={'Quinones':"tabor_nosulf.dump",
              'Wang':'wang_data.dump',
              'Gecko':'gecko_full.dump',
              'ExpMoNA':"expmona_cleaned.dump",
              'nablaDFT':"nablaDFT.dump",
              'Massbank, EU':"massbankeu.dump"}
    datasets={'Gecko':'gecko_full.dump','Wang':'wang_data.dump', 'Quinones':"tabor_nosulf.dump"}
    #datasets={'nablaDFT':"nablaDFT.dump",'ExpMoNA':"expmona_cleaned.dump", 'Massbank, EU':"massbankeu.dump"}
    return datasets

def plot_number_of_heavy_atoms(dataframe, label, foldername):
    "Plot the distribution of the heavy atoms"
    sns.color_palette("colorblind")
    plt.rc('font', size=8)
    dataframe['# heavy atoms'] = dataframe['# heavy atoms'].astype('float32')
    fig, ax1 = plt.subplots(figsize=(2.38, 2.0), dpi=300, constrained_layout=True)
    sns.histplot(dataframe, x='# heavy atoms', ax=ax1, bins=25, discrete=True)
    plt.ylabel('Number of molecules')
    plt.xlabel('Number of heavy atoms')
    ax1.set_xticks(np.arange(0, np.max(dataframe['# heavy atoms'])+1, 10))
    plt.setp(ax1.get_xticklabels(), rotation=30, horizontalalignment='right')

    plt.tight_layout()
    plt.savefig(foldername+'/'+label+'_heavyAtoms_dist.png', bbox_inches='tight', dpi=320)
    plt.close()
    
def plot_number_of_heavy_atoms(dataframe, label, foldername):
    "Plot the distribution of the heavy atoms"
    sns.color_palette("colorblind")
    plt.rc('font', size=8)
    dataframe['# heavy atoms'] = dataframe['# heavy atoms'].astype('float32')
    fig, ax1 = plt.subplots(figsize=(2.38, 2.0), dpi=300, constrained_layout=True)
    sns.histplot(dataframe, x='# heavy atoms', ax=ax1, bins=25, discrete=True)
    plt.ylabel('Number of molecules')
    plt.xlabel('Number of heavy atoms')
    ax1.set_xticks(np.arange(0, np.max(dataframe['# heavy atoms'])+1,np.round(np.max(dataframe['# heavy atoms'])/10)))
    plt.setp(ax1.get_xticklabels(), rotation=30, horizontalalignment='right')

    plt.tight_layout()
    plt.savefig(foldername+'/'+label+'_heavyAtoms_dist.png', bbox_inches='tight', dpi=320)
    plt.close()
    
def arange_heavy_atoms(dataframe, label):
    a_r = pd.concat([dataframe['# heavy atoms'].astype('float32').reset_index(drop=True), 
               pd.DataFrame(len(dataframe['# heavy atoms'])*[label]).rename({0:'label'}, axis='columns')], axis=1)
    return a_r

def pie_atom_plot(orig, foldername, title): 
    """plotting piechart of atom types"""

    plt.rc('font', size=8)
    colors = sns.color_palette("colorblind")[0:40]
    fig, ax1 = plt.subplots(figsize=(2.38, 2.0), dpi=300,
                            constrained_layout=True)
    orig2 = orig.filter(regex='Atoms').copy(deep=True)
    orig2.columns = orig2.iloc[0]
    orig2.drop(0, inplace=True)
    orig2 = orig2[orig2.columns[orig2.columns != 'H']].astype('float32')
    orig2 = pd.DataFrame(orig2.sum()/orig2.sum().sum())
    orig2.to_csv(foldername+'/'+label+'atomtypes')
    other = orig2.loc[orig2.values < 0.01].index.values
    orig2 = pd.concat([orig2.loc[orig2.values > 0.01], 
           pd.DataFrame({'other':orig2.loc[orig2.values < 0.01].sum().values}).transpose()])
    plt.pie(np.concatenate(orig2.values, axis=0), 
            labels = orig2.index.values, colors=colors, autopct='%.0f%%')
    orig2.to_csv(foldername+'/'+title)
    plt.tight_layout()
    plt.savefig(foldername+'/'+title+'.png',
            bbox_inches='tight', dpi=320)
    plt.close()
```

```python
foldername = '../reports/figures/'
output_filepath = '../data/output/'
```

```python
datasets = load_datasets()
ar_list = []
adist_list = []
at_list = []

for label in datasets.keys(): 
    d=pd.read_csv(output_filepath+label+'.descr.csv', 
                  compression={'method':'gzip', 'compresslevel': 1,'mtime': 1}, 
                  low_memory=False)
    orig = d.copy(deep=True)
    pie_atom_plot(d, foldername, label+'_pie_plot')
    a_types = arange_atomic_distribution(d).reset_index().rename({0:'number'}, axis='columns')
    at_list.append(pd.concat([a_types,
                          pd.DataFrame(len(a_types)*[label]).rename({0:'label'}, axis='columns')], axis=1))
    d.columns = d.iloc[0]
    d.drop(0, inplace=True)
    a_r = arange_atomic_ratios(d, label)
    ar_list.append(a_r)  
    adist_list.append(arange_heavy_atoms(d, label))
    plot_number_of_heavy_atoms(d, label, foldername)
    
at_pd = pd.concat(at_list, axis=0)   
ar_pd = pd.concat(ar_list, axis=0)    
adist_pd = pd.concat(adist_list, axis=0)  
```
